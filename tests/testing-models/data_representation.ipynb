### Test results 

Poniższy notebook zawiera analzę wyników uzyskanych z implementacji filtrów molekularnych na wybranych zbiorach danych: 
- BACE,
- BBBP
- ClinTox
- HIV


```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=Warning)
```

```python
df = pd.read_json("filter_results.json")
df
```

```python
df
```

```python
top_3_per_group = {
    name: group.nlargest(3, 'filter_difference')
    for name, group in df.groupby('dataset_name')
}
```

```python
for name, group in top_3_per_group.items():
    print(f"Dataset '{name}':")
    print(group)
    print("-" * 40)
```
